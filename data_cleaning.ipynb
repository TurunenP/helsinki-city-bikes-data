# Importing Required Libraries in python

````python
# imports

import numpy as np
import pandas as pd
````

### Geting Data from the url Given

````python
# load data

data_url = "https://dev.hsl.fi/citybikes/od-trips-2021/2021-05.csv"

df = pd.read_csv(data_url)
````

### Geting the head of the dataframe

````python
df.head()
````

### Geting the shape of the dataframe

````python
df.shape
````

### Using ```df.drop``` to drop Duration column with less than 10 sec

````python
df = df.drop(df[df['Duration (sec.)'] < 10].index)
````

### Using ```df.drop``` to drop Covered Distance column with less than 10 meters

````python
df = df.drop(df[df['Covered distance (m)'] < 10].index)
````

### Printing out the dataframe after droping the columns

````python
df
````

````python
df
````

### Saving the cleaned Data in csv using ``` df.to_csv()``` 

````python
df.to_csv('filtered_data.csv', index=False)
````
